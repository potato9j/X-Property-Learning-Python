# TransformerмҷҖ м–ём–ҙлӘЁлҚёл§Ғ
мӢӨм ң нҷҳмһҗ м •ліҙк°Җ м•„лӢҢ н•©м„ұ лҚ°мқҙн„°лЎң мӢӨн–үн•©лӢҲлӢӨ.


## мӢӨн–ү м „ мҳҲмёЎ
м¶ңл Ҙкіј shapeлҘј лЁјм Җ мҳҲмғҒн•©лӢҲлӢӨ.


```python
tokens = ["нҷҳмһҗ", "нҳҲм••", "мғҒмҠ№", "нҷ•мқё"]
token_to_id = {token: index for index, token in enumerate(tokens)}
input_ids = [token_to_id[token] for token in tokens]
attention_mask = [1 for _ in input_ids]
print("tokens:", tokens)
print("input_ids:", input_ids)
print("attention_mask:", attention_mask)
print("shape-like:", (1, len(input_ids)))
```
