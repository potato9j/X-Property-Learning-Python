# 조건문, 반복문과 comprehension
실제 환자 정보가 아닌 합성 데이터로 실행합니다.


## 실행 전 예측
출력과 shape를 먼저 예상합니다.


```python
# 실습용 시작 코드입니다. 값을 바꾸며 결과를 비교하세요.
day = 4
topic = "조건문, 반복문과 comprehension"
scores = [72, 85, 91, 68, 77]
passed = []
for score in scores:
    if score >= 75:
        passed.append(score)
summary = {"day": day, "topic": topic, "student_count": len(scores), "passed_count": len(passed), "average": sum(scores) / len(scores)}
print(summary)
print(type(summary["average"]).__name__)
reflection = "예측과 실행 결과를 비교해 적습니다."
print(reflection)
```
